```python
# Convert the  dictionary to an object

def dict_to_obj(d):
    if isinstance(d, dict):
        obj = type('obj', (object,), {})
        for k, v in d.items():
            k = k.replace('.', '_')
            setattr(obj, k, dict_to_obj(v))
        return obj
    else:
        return d

    
loaded_data_obj = dict_to_obj(loaded_data_dict)

# Access the loaded data using dot notation
loaded_data_obj.countCDR3_L.TRA.L_10


```
