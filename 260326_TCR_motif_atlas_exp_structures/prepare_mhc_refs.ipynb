```python
import pandas as pd
import numpy as np
import glob
import os
```

```python
topdir = '/Users/roessner/Documents/PostDoc/Data/TCR_data/mhc_alleles'

# mhci
df = pd.read_csv(f'{topdir}/data_classI_MHC_I_sequences.txt', sep='\t')

# most common alleles
df_common = pd.read_csv('tmp/alleles_mhci.txt', sep = '\t')
common_alleles = np.unique(df_common['Allele'])
common_alleles = np.append(common_alleles, 'B3505')
df = df[df['Allele'].isin(common_alleles)]

df = df.sort_values(by='Allele')
df.to_csv(f'{topdir}/data_classI_MHC_I_sequences_reduced.txt', sep='\t', index=False)
```

```python
topdir = '/Users/roessner/Documents/PostDoc/Data/TCR_data/mhc_alleles'

# mhcii
df = pd.read_csv(f'{topdir}/data_classII_MHC_II_sequences.txt', sep='\t')
df['Allele'] = df['Allele'].str.replace('*', '', regex=False).str.replace(':', '', regex=False)

# most common alleles
df_common = pd.read_csv('tmp/alleles_mhcii.txt', sep = '\t')
common_alleles = np.unique(df_common['Allele'])

common_alleles_new = []
for a in common_alleles:
    common_alleles_new.extend(a.split('__'))
common_alleles_new = np.unique(common_alleles_new)
common_alleles_new = [x for x in common_alleles_new if not x.startswith(('BoLA', 'Gaga'))]
common_alleles_new = [x.replace('_', '') for x in common_alleles_new]

df = df[df['Allele'].isin(common_alleles_new)]
df.to_csv(f'{topdir}/data_classII_MHC_II_sequences_reduced.txt', sep='\t', index=False)
```
