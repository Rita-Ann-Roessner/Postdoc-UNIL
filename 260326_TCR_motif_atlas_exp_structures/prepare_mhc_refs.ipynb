```python
import pandas as pd
import numpy as np
```

```python
topdir = '/Users/roessner/Documents/PostDoc/Data/TCR_data/mhc_alleles'

# get alleles used in tcr motif atlas
df = pd.read_csv(f'{topdir}/All_models_TCRs.csv')
motif_alleles = np.unique(df['MHC'])
motif_alleles = motif_alleles[motif_alleles != 'na']
motif_alleles = [item for x in motif_alleles for item in x.split('-')]

# mhci
df = pd.read_csv(f'{topdir}/data_classI_MHC_I_sequences.txt', sep='\t')

dico = {
    'Allele' : [],
    'Sequence' : [],
}

for seq, group in df.groupby('Sequence'):
    dico['Sequence'].append(seq)

    alleles = np.unique(group['Allele'])
    alleles.sort()


    if len(alleles) == 1:
        dico['Allele'].append(alleles[0])
    else:
        # alleles that are representend in the motif atlas 
        allele_motif = [x for x in alleles if x in motif_alleles]
        if len(allele_motif) == 0:
            dico['Allele'].append(alleles[0])
        else:
            dico['Allele'].append(allele_motif[0])


df = pd.DataFrame(dico)
df = df.sort_values(by='Allele')
df.to_csv(f'{topdir}/data_classI_MHC_I_sequences_reduced.txt', sep='\t', index=False)
```

```python
topdir = '/Users/roessner/Documents/PostDoc/Data/TCR_data/mhc_alleles'

# get alleles used in tcr motif atlas
df = pd.read_csv(f'{topdir}/All_models_TCRs.csv')
motif_alleles = np.unique(df['MHC'])
motif_alleles = motif_alleles[motif_alleles != 'na']
motif_alleles = [item for x in motif_alleles for item in x.split('-')]

# mhcii
df = pd.read_csv(f'{topdir}/data_classII_MHC_II_sequences.txt', sep='\t')

dico = {
    'Allele' : [],
    'Sequence' : [],
}

for seq, group in df.groupby('Sequence'):
    dico['Sequence'].append(seq)

    alleles = np.unique(group['Allele'])
    alleles.sort()

    # clean allele notation
    alleles = np.array([str(x).replace('*', '').replace(':', '') for x in alleles])

    if len(alleles) == 1:
        dico['Allele'].append(alleles[0])
    else:
        # alleles that are representend in the motif atlas 
        allele_motif = [x for x in alleles if x in motif_alleles]
        if len(allele_motif) == 0:
            dico['Allele'].append(alleles[0])
        else:
            dico['Allele'].append(allele_motif[0])

df = pd.DataFrame(dico)
df = df.sort_values(by='Allele')
df = df[~df['Allele'].str.startswith('BoLA')] # remove bovine
df = df[~df['Allele'].str.startswith('Gaga')] # remove chicken 
df.to_csv(f'{topdir}/data_classII_MHC_II_sequences_reduced.txt', sep='\t', index=False)
```
